```python
import pandas as pd 
import networkx as nx
```

```python
drug_graph = nx.read_gexf("../data/drug_association_graph.gexf")
```

```python
drugs = {}
conditions = {}
for k , v in drug_graph.nodes(data=True):
    if v['type'] == 'drug':
        drugs[k] = v
    else:
        conditions[k] = v
```
